```python
from scrapeghost import SchemaScraper

url = "https://mediapartychicago2023.sched.com/"
schema =  [{"date": "", "event_name": "", "speaker_name": "", "time": "", "location": ""}]
scraper = SchemaScraper(schema)

scraper.scrape(url)
```

```python
from scrapeghost import CSS

scraper = SchemaScraper(
    schema,
    extra_preprocessors=[
        CSS("#sched-content"),
    ]
)
scraper.scrape(url)

```

```python
scraper = SchemaScraper(
    schema,
    extra_preprocessors=[
        CSS(".sched-container, .sched-container-dates"),
    ],
    auto_split_length=3000,
)
scraper.scrape(url)
```

```python
scraper = SchemaScraper(
    schema,
    extra_preprocessors=[
        CSS(".sched-container, .sched-container-dates"),
    ],
    auto_split_length=4000,
    models=["gpt-4"],
    extra_instructions=["The date is in an <h3> tag."],
)
scraper.scrape(url)
scraper.scrape
```
